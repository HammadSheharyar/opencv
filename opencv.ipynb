```python
import numpy as np
custom_image = np.zeros((512,512, 3))
rectangle = cv2.rectangle(custom_image, pt1= (200,100), pt2 = (400,300), color=(255, 0, 0), thickness = 3)
cv2.imshow("window", custom_image)
cv2.waitKey(0)
```

```python
# circle
cv2.circle(custom_image, center = (100,300), radius = (50), color = (0, 0, 255), thickness = 3)
cv2.imshow("window",custom_image)
cv2.waitKey(0)
```

```python
# line
cv2.line(custom_image, pt1= (0,0), pt2=(400,0), color =(0,255,0), thickness= 3)
cv2.imshow("window", custom_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# text
img2 = np.ones((512,512,3))
cv2.putText(img2,text = "Hello,how are you",fontFace = cv2.FONT_ITALIC, org = (100,100), color = (255,0,0),fontScale = 1, thickness = 2, lineType = cv2.LINE_AA)
cv2.imshow("window", img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# events: clicking image to produce circle etc
img = np.zeros((512,512, 3))

while True:
    cv2.imshow("window", img)
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
cv2.destroyAllWindows()
```

```python
# events: clicking image to produce circle etc
import cv2
import numpy as np
img = np.zeros((512,512, 3))
def draw(event, x, y ,flag,params ):
    if event == 1:
        cv2.circle(img, center = (x, y), radius = 60, color = (255, 0 , 0), thickness = -1)
    # print("Event triggered")


cv2.namedWindow(winname= "window")
cv2.setMouseCallback("window", draw)
while True:

    cv2.imshow("window", img)
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
cv2.destroyAllWindows()
```
